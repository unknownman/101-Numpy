

$$
\vec{a} = \begin{pmatrix}
a_1 \\
a_2 \\
\vdots \\
a_n
\end{pmatrix}
$$

$$
\vec{b} = \begin{pmatrix}
b_1 \\
b_2 \\
\vdots \\
b_n
\end{pmatrix}
$$

$$
d(a, b) = 1 - \frac{a \cdot b}{\|a\| \|b\|}
$$

<div style="direction:rtl; font-family: Portada ARA">
    
-   a · b: ضرب داخلی دو بردار a و b
-   |a|: نرم (طول) بردار a
-   |b|: نرم (طول) بردار b

</div>

```python
import numpy as np

def cosine_distance(a, b):
  dot_product = np.dot(a, b)
  norm_a = np.linalg.norm(a)
  norm_b = np.linalg.norm(b)
  return 1 - (dot_product / (norm_a * norm_b))

vector1 = np.array([1, 2, 3])
vector2 = np.array([4, 5, 6])

distance = cosine_distance(vector1, vector2)
print("Cosine Distance:", distance)

vectors = np.array([[1, 2], [4, 5], [7, 8]])
distances = 1 - np.dot(vectors, vectors[0]) / (np.linalg.norm(vectors, axis=1) * np.linalg.norm(vectors[0]))
print("Distances:", distances)







```
