```python
import numpy as np
import matplotlib.pyplot as plt

data = np.array([
    [150, 3, 2],  
    [120, 2, 1],
    [200, 4, 3],
    [180, 3, 2],
    [100, 2, 1]
])

feature_space = data

print("Feature Space:\n", feature_space)

plt.scatter(feature_space[:, 0], feature_space[:, 1])
plt.xlabel("Area (sq m)")
plt.ylabel("Number of Rooms")
plt.title("2D Feature Space")
plt.grid(True)
plt.show()
```
