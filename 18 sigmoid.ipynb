$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$


<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/8/88/Logistic-curve.svg/2560px-Logistic-curve.svg.png" />


```python
import numpy as np
import matplotlib.pyplot as plt

# تعریف تابع سیگموید
def sigmoid(x):
  return 1 / (1 + np.exp(-x))

# ایجاد آرایه ورودی
x = np.linspace(-10, 10, 100) # 100 عدد بین -10 و 10

# محاسبه خروجی تابع سیگموید
y = sigmoid(x)

# رسم نمودار
plt.plot(x, y)
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.show()

# نمایش چند نمونه از خروجی تابع
print(sigmoid(np.array([-1, 0, 1, 10]))) # خروجی برای ورودی های مختلف
```
