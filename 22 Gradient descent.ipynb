

$$
\theta := \theta - \alpha \nabla J(\theta)
$$

<div style="direction:rtl; font-family: Portada ARA; margin:0 ; padding:0">
    
- θ: پارامترهای مدل
- α: نرخ یادگیری (Learning Rate)
- ∇J(θ): گرادیان تابع هزینه

</div>


```python
import numpy as np

def cost_function(X, y, theta):
  m = len(y)
  predictions = X @ theta
  errors = predictions - y
  cost = np.sum(errors**2) / (2 * m)
  return cost

def gradient_descent(X, y, theta, alpha, num_iters):
  m = len(y)
  for _ in range(num_iters):
    predictions = X @ theta
    errors = predictions - y
    gradient = X.T @ errors / m
    theta = theta - alpha * gradient
  return theta

X = np.array([[1, 2], [1, 3], [1, 4]])
y = np.array([3, 5, 6])
theta = np.array([0, 0])
alpha = 0.01
num_iters = 1000

theta = gradient_descent(X, y, theta, alpha, num_iters)
print("Theta:", theta)
```
