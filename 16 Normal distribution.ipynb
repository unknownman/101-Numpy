### <div dir="rtl">مفهوم ریاضی توزیع پواسون:
</div>


$$
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
$$

----
$$
P(X = 3) = \frac{5^3 e^{-5}}{3!} \approx \frac{125 \times 0.0067}{6} \approx 0.14
$$





### <div dir="rtl" style="padding-top:100px">پیاده‌سازی توزیع پواسون با Numpy: </div>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson


# پارامتر توزیع پواسون
lam = 5       # نرخ متوسط رخدادها
size = 1000   # تعداد نمونه‌ها

data = np.random.poisson(lam, size)

plt.hist(data, bins=np.arange(0, max(data)+2)-0.5, density=True, alpha=0.6, color='orange', rwidth=0.8)
plt.xticks(range(max(data)+1))

plt.title('Poisson Distribution')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.show()
```

```python
x = np.arange(0, max(data)+1)
probs = poisson.pmf(x, lam)
plt.plot(x, probs, 'bo', ms=8)
plt.vlines(x, 0, probs, lw=2, colors='b')
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.title("Probability Mass Function")
plt.show()
```
