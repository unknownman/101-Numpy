<font size="12">
$$
\int_a^b f(x) \, dx
$$
</font>







```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

a = 0 
b = 1
n = 100

x = np.linspace(a,b,n)
 
y = f(x)
integral = np.trapezoid(y,x)
print("integral : ", integral)

plt.plot(x,y)
plt.fill_between(x,y,color="skyblue", alpha=0.5)
plt.title("Integral Trapezoidal Rule")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```
